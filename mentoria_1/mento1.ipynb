```python
#######################################################
# Elegimos estas 5 variables para realizar el analisis
#######################################################
# IMC    -> Es el indice de masa corporal, contamos con una version categoria clasifIMC.
# cc     -> Circunferencia de cintura, se anota en cm.
# mets   -> Nivel de actividad fisica.
# fgr2   -> Consumo de grasa en gramos.
# tenmax -> Promedio de las tensiones arteriales maximas.
elegidas      = ['IMC','cc','mets','fgr2']
nuevo_dataset = obedf[elegidas]
nuevo_dataset = nuevo_dataset.join( pd.DataFrame( data = (obedf.ten1max + obedf.ten2max) / 2, columns=['tenmax']) )
elegidas.append('tenmax')

#######################################################
# Ademas elegimos otro subset de variables que pueden ser utilies en el analisis, junto con algunas auxiliares
#######################################################
# clasifIMC -> //[FIXME] Poner explicacion.
# riesgo    -> Riesgo de enfermedad cardiovascular y metabólica según circunferencia, cambia de acuerdo al genero.
# mets_cat  -> De 0 a 599 - sedentario, de 600 a 1490 - activo, mas de 1500 - muy activo.
# fgr1      -> Ingesta segun frecuencia 1- bajo, 2-adecuado, 3-alto.
# hta       -> Presencia de presion arterial 1-no, 2-si.
categoricas   = ['clasifIMC','riesgo','fgr1']

# Creamos la categoria para mets
ranges        = [0 - 0.1, 599 , 1499 , nuevo_dataset.mets.max() + 0.1]
labels        = ['Sedentario', 'Activo', 'Muy Activo']
mets_cat      = pd.cut(nuevo_dataset.mets, ranges, right=False, labels=labels)
nuevo_dataset = nuevo_dataset.join( pd.DataFrame( data={'mets_cat' : mets_cat}) )

# Agregamos
auxiliares    = ['sexo','edad','hta']
nuevo_dataset = nuevo_dataset.join( obedf[categoricas+auxiliares] )

#######################################################
# Y con eso definimos un nuevo dataset mas reducido con esas variables
#######################################################
nuevo_dataset.shape
```

```python
#######################################################
# Calculamos los estadisticos clasicos
#######################################################
describe_data   = pd.DataFrame (data = nuevo_dataset[elegidas].describe().round(2))

mode_dataframe  = pd.DataFrame (data = nuevo_dataset[elegidas].mode().round(2)) 
rename_dict     = {}
for mode_index in range(len(mode_dataframe.index)):
  rename_dict.update( {mode_index : 'mode_'+ str(mode_index)} )
mode_dataframe  = mode_dataframe.rename(index = rename_dict)

describe_data   = describe_data.append( mode_dataframe )
```

```python
# Definimos una funcion para facilitar el analisis de las variables
plt.figure(figsize=(10,5))
def print_and_graph(continious, categoric, categoric_desc):
  print ('*** Estadisticos Clasicos ***')
  print (describe_data[continious].dropna())

  print('*** Test de Shapiro-Wiks ***')
  test = scipy.stats.shapiro(nuevo_dataset[continious])
  print('Estadistico: ' + str(test[0]))
  print('p_valor: ' + str(test[1]))

  fig, ax = plt.subplots(2,2)

  dataset_hombre = nuevo_dataset[nuevo_dataset.sexo == 1]
  dataset_mujer  = nuevo_dataset[nuevo_dataset.sexo == 2]

  seaborn.set(style="whitegrid")
  seaborn.distplot (nuevo_dataset[continious] , ax=ax[0][0], color='#1f93bc')

  seaborn.distplot (dataset_hombre[continious] , ax=ax[1][0])
  seaborn.distplot (dataset_mujer [continious] , ax=ax[1][0])

  seaborn.countplot(data=nuevo_dataset, x=categoric, ax=ax[0][1], color='#1f93bc')
  seaborn.countplot(data=nuevo_dataset, x=categoric, ax=ax[1][1], hue='sexo')
  
  ax[1][1].set_xticklabels( categoric_desc )
  ax[0][1].set_xticklabels( categoric_desc )
  ax[1][1].legend(['Hombre','Mujer'])
  ax[1][0].legend(['Hombre','Mujer'])

  ax[0][0].title.set_text( 'A' )
  ax[0][1].title.set_text( 'B' )
  ax[1][0].title.set_text( 'C' )
  ax[1][1].title.set_text( 'D' )

  fig.tight_layout(pad = 0.2)
  fig.set_figheight(10)
  fig.set_figwidth(14)
  
  fig.show()
```

```python
description_0 = ['Peso\nBajo','Peso\nnormal','Preobesidad','Obesidad\ngrado I','Obesidad\ngrado II','Obesidad\ngrado III']
description_1 = ['Menor de 18.5 kg/m2','18.5 a 24.9 kg/m2','25 a 29.9 kg/m2','30 a 34.9 kg/m2','35 a 39.9 kg/m2','Mayor a 40 kg/m2']

print_and_graph('IMC','clasifIMC', description_0)
```

```python
from scipy.stats import kendalltau
def to_cuartiles(data, column):   
  cuartiles = [data[column].describe()[i] for i in range(8) if i > 2]   # generar los intervalos para la discretización de la edad en Q1 a Q4
  data[column+'_Q'] = pd.cut(data[column], cuartiles, labels=[0, 1, 2, 3])
```

```python
var_to_cuartiles = ['tenmax', 'IMC', 'cc', 'fgr2', 'edad']
for i in range(len(var_to_cuartiles)):
  to_cuartiles(nuevo_dataset, var_to_cuartiles[i])
```

```python
var_eleg = ['tenmax_Q', 'IMC_Q', 'cc_Q', 'fgr2_Q','edad_Q'] 
var_names = ['tensión máxima promedio', 'índice de masa corporal', 
             'circunferencia de cintura', 'consumo de grasa en gramos por día', 'edad']

edad_labels    = ['18-26' , '26-41'  , '41-57'  , '57-97'  ]
ten_max_labels = ['90-115', '115-120', '120-126', '126-170']
IMC_labels     = ['14-22' , '22-25'  , '25-28'  , '28-51'  ]
cc_labels      = ['52-75' , '75-85'  , '85-96'  , '96-196' ]
fgr2_labels    = ['0-19'  , '19-27'  , '27-38'  , '38-528' ]

var_labels = [ten_max_labels, IMC_labels, cc_labels, fgr2_labels, edad_labels]

nuevo_dataset_d = nuevo_dataset[var_eleg]
nuevo_dataset_d = nuevo_dataset_d.dropna()

for i in range(len(var_eleg)):
  for j in range(len(var_eleg)):
    if i != j:
      if i < j:
        plt.figure(figsize=(10,6))

        exp = pd.crosstab(nuevo_dataset[var_eleg[i]], nuevo_dataset[var_eleg[j]], normalize=True).round(2)
        ax = seaborn.heatmap(exp, cmap='Blues', xticklabels=var_labels[j], yticklabels= var_labels[i])

        plt.xlabel(var_names[j], fontsize=15)
        plt.ylabel(var_names[i], fontsize=15)
        plt.title('\n\n'+var_names[j]+' vs '+var_names[i]+'\n\n', fontsize=20)

        ax.invert_yaxis()

        plt.show()

        a_kendall = kendalltau(nuevo_dataset_d[var_eleg[j]], nuevo_dataset_d[var_eleg[i]])

        print('\n\ncorrelación de kendall en :'+var_names[j]+' vs '+var_names[i]+' : ', a_kendall[0])
        print('p-valor: ', a_kendall[1], '\n\n')
```
